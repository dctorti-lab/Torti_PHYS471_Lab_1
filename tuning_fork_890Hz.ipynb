Here we wish to plot our FFT directly from the oscilloscope. I will be utilizing the same code that I employed in Advanced Lab 1 (PHYS 371) for my FFT plots, as well as some extra graphical pizzazz taken from Professor Hedberg's sample FFT plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
filepath = "/content/drive/MyDrive/tuning_fork_890Hz.csv"
data = np.genfromtxt(filepath, delimiter=',', names = ["C1","C2"])
```

```python
frequency = data['C1']
level = data['C2']
```

```python
fig, ax = plt.subplots()
ax.plot(frequency, level, color="lightblue", marker='o', mec='steelblue',mfc='steelblue',ms='3')
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Intensity (dB)')
ax.set_title('Intensity vs Frequency')
ax.grid
plt.show()
```
